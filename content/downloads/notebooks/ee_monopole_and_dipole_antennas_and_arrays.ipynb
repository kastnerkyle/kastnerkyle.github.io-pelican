```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Elevation points
n_theta = 30 
#Azimuth points
n_phi = 50
theta = 2 * np.pi * np.linspace(0., 1., n_theta)
phi = 2 * np.pi * np.linspace(0., 1., n_phi)
[theta, phi] = np.meshgrid(theta, phi)
```

```python
#Halfwave Dipole

B = 0.5
num = np.cos(np.pi * B * np.cos(theta)) - np.cos(np.pi * B)
denom = np.sin(theta)
resp = num / denom
 
X = resp * np.sin(theta) * np.cos(phi)
Y = resp * np.sin(theta) * np.sin(phi)
Z = resp * np.cos(theta)
 
fig = plt.figure()
ax = Axes3D(fig)
ax._axis3don = False
ax.plot_wireframe(X, Y, Z, color='steelblue')
ax.plot([0, 0], [0, 0], [np.nanmin(Z), np.nanmax(Z)], linewidth=3.0, color='darkred')
```

```python
#Quarterwave Monopole
B = 0.25
num = np.cos(np.pi * B / 2. * np.cos(theta)) - np.cos(np.pi * B / 2.)
denom = np.sin(theta)
resp = num / denom
 
X = resp * np.sin(theta) * np.cos(phi)
Y = resp * np.sin(theta) * np.sin(phi)
Z = resp * np.cos(theta)

#Groundplane hack
Z[np.where(Z < 0)] = 0.

fig = plt.figure()
ax = Axes3D(fig)
ax._axis3don = False

ax.plot_wireframe(X, Y, Z, color='steelblue')
ax.plot([0, 0], [0, 0], [np.nanmin(Z), np.nanmax(Z)], linewidth=3.0, color='darkred')
```
