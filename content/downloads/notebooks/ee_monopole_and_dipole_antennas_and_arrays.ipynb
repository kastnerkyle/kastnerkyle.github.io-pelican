```python
%pylab inline
#Dipole response
freq = 1
wavelength = 1. / (freq)
n_points = 100
l = .5
r = np.linspace(-0.5,.5,n_points)
phi = 2 * np.pi * r
B = 2 * np.pi / wavelength
f1 = np.cos(B * l / 2. * np.cos(phi))
f2 = np.cos(B * l / 2.)
f3 = np.sin(phi)
E = (f1 - f2) / f3
En = np.abs(E)

ax = plt.subplot(111, polar=True)
ax.plot(phi, En, color='r', linewidth=3)
ax.grid(True)
plt.show()
```

```python
r = np.linspace(-0.5,.5,n_points)
theta = 2 * np.pi * r
Hn = np.array([1.]*r.shape[0])

ax = plt.subplot(111, polar=True)
ax.plot(theta, Hn, color='r', linewidth=3)
ax.grid(True)
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

X, Y = np.meshgrid(theta, phi)
Z = Hn * En * np.sin(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
```

```python
#Monopole response
freq = 1
wavelength = 1. / (freq)
l = .25
B = 2 * np.pi / wavelength
r = np.linspace(-0.25,.25,1000)
theta = 2 * np.pi * r
f1 = np.cos(B * l / 2. * np.cos(theta))
f2 = np.cos(B * l / 2.)
f3 = np.sin(theta)
E = (f1 - f2) / f3
En = np.abs(E)
ax = plt.subplot(111, polar=True)
ax.plot(theta, En, color='r', linewidth=3)
ax.grid(True)
plt.show()
```
