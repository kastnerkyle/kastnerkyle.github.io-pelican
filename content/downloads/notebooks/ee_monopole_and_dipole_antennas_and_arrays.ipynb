```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Elevation points
n_theta = 30 
#Azimuth points
n_phi = 50
theta = 2 * np.pi * np.linspace(0., 1., n_theta)
phi = 2 * np.pi * np.linspace(0., 1., n_phi)
[theta, phi] = np.meshgrid(theta, phi)
```

```python
#Halfwave Dipole
antennaX = [0, 0]
antennaY = [0, 0]

B = 0.5
num = np.cos(np.pi * B * np.cos(theta)) - np.cos(np.pi * B)
denom = np.sin(theta)
resp = num / denom
 
X = resp * np.sin(theta) * np.cos(phi)
Y = resp * np.sin(theta) * np.sin(phi)
Z = resp * np.cos(theta)
 
fig = plt.figure()
ax = Axes3D(fig)
ax._axis3don = False
ax.plot_wireframe(X, Y, Z)
ax.plot(antennaX, antennaY, [np.nanmin(Z), np.nanmax(Z)], linewidth=3.0, color='red')
```

```python
#Quarterwave Monopole
B = 0.25
num = np.cos(np.pi * B / 2. * np.cos(theta)) - np.cos(np.pi * B / 2.)
denom = np.sin(theta)
resp = num / denom
 
X = resp * np.sin(theta) * np.cos(phi)
Y = resp * np.sin(theta) * np.sin(phi)
Z = resp * np.cos(theta)

#Groundplane hack
Z[np.where(Z < 0)] = 0.

fig = plt.figure()
ax = Axes3D(fig)
ax._axis3don = False

ax.plot_wireframe(X, Y, Z)
ax.plot(antennaX, antennaY, [np.nanmin(Z), np.nanmax(Z)], linewidth=3.0, color='red')
```

A generalized response from a number of monopole elements

```python
class BaseElement:
    def __init__(self):
        pass
    
    def _get_propagation(self, az, el):
        raise AssertionError("This class is abstract - antenna elements should override this method!")

class MonopoleElement(BaseElement):
    def _get_propagation(self, az, el=np.pi / 2):
        return np.array([np.cos(az) * np.sin(el), np.sin(az) * np.sin(el), np.cos(el)])[:, None]    

class GeometryMixin:
    def _get_geometry(self, n_ant, geometry='linear', wavelength_spacing=.5, random_seed=None):
        if geometry == 'linear':
            return np.array([[i, 0, 0] for i in range(n_ant)]), wavelength_spacing
        elif geometry == 'random':
            random_state = np.random.RandomState(random_seed)
            return random_state.rand(n_ant, 3), wavelength_spacing
        else:
            raise ValueError('Only linear arrays are currently supported')
    
    def _center_min_max(self, coords):
        return np.mean(coords), np.min(coords), np.max(coords)
    
    def plot_geometry(self, geometry=None):
        if geometry is None:
            geometry = self.geometry
        self.x_coords = geometry[:, 0]
        self.y_coords = geometry[:, 1]
        self.x_center, self.x_min, self.x_max = self._center_min_max(self.x_coords)
        self.y_center, self.y_min, self.y_max = self._center_min_max(self.y_coords)
        plt.plot(self.x_coords, self.y_coords, 'o', color='steelblue')
        plt.xlim(self.x_min - 1, self.x_max + 1)
        plt.ylim(self.y_min - 1, self.y_max + 1)
        
class ArrayMixin:
     def __init__(self, n_ant):
        self.n_ant = n_ant
        self.geometry, self.wavelength_spacing = self._get_geometry(self.n_ant, 'linear', .5)
        self.beam_weights = self._get_beam_weights()
    
     def _get_beam_weights(self):
        return np.matrix([1.0] * self.n_ant).T
        
     def gain_response(self, az_arr, el_arr=None):
        az_gains = np.zeros(len(az_arr)) 
        for n, az in enumerate(az_arr):
            propagation = p = self._get_propagation(az)
            response = a = np.matrix(np.exp(-2j * np.pi * np.dot(self.geometry, p) * self.wavelength_spacing))
            az_gains[n] = np.abs(np.dot(self.beam_weights.H, a))[0, 0] #had to make a matrix for easy hermitian
        return az_gains
        
     def _plot_beam_quadrant(self, quadrant, n_pts, x_plot_min, x_plot_max, y_plot_min, y_plot_max):
         all_az = np.linspace(-np.pi, np.pi, n_pts)
         all_gain = self.gain_response(all_az)
         self.min_gain = np.min(all_gain)
         self.max_gain = np.max(all_gain)
         az_spacing = all_az[1] - all_az[0]
         if quadrant in [1, 3]:
             xs = np.linspace(x_plot_min, self.x_center, n_pts)
         elif quadrant in [2, 4]:
             xs = np.linspace(self.x_center, x_plot_max, n_pts)
         else:
             raise ValueError("Quadrant must be 1, 2, 3 or 4")
    
         for az in all_az:
             y1s = xs*tan(az - .5 * az_spacing)
             y2s = xs*tan(az + .5 * az_spacing)
             if quadrant in [1, 2]:
                 plt.fill_between(xs, y1=y1s, y2=y2s, where=(y1s >= -.5 * az_spacing) & (y2s >= -.5 * az_spacing))
             elif quadrant in [3, 4]:
                 plt.fill_between(xs, y1=y1s, y2=y2s, where=(y1s <= .5 * az_spacing) & (y2s <= .5 * az_spacing))
    
     def plot_beam(self, n_pts=20):
         self.plot_geometry()
         x_plot_min = self.x_min - 1000
         x_plot_max = self.x_max + 1000
         y_plot_min = self.y_min - 1000
         y_plot_max = self.y_max + 1000
         plt.xlim(x_plot_min, x_plot_max)
         plt.ylim(y_plot_min, y_plot_max)
         for i in [1,2,3,4]:
             self._plot_beam_quadrant(i, n_pts, x_plot_min, x_plot_max, y_plot_min, y_plot_max)
             break
        
     def plot_gain(self, n_pts=50, min_az=-np.pi, max_az=np.pi, log_scale=True):
         all_gain = self.gain_response(np.linspace(min_az, max_az, n_pts))
         if log_scale is True:
             all_gain = 10 * np.log(all_gain)
         plt.plot(all_gain, color='steelblue')
    
class ClassicalBeamformerMixin(ArrayMixin):
     def __init__(self, n_ant, beam_dir):
        self.beam_dir = beam_dir
        ArrayMixin.__init__(self, n_ant)
        self.geometry, self.wavelength_spacing = self._get_geometry(self.n_ant, 'linear', .5)
        
     def _get_beam_weights(self):
         p = propagation = self._get_propagation(self.beam_dir)
         a = response = np.matrix(np.exp(-2j * np.pi * np.dot(self.geometry, p) * self.wavelength_spacing))
         return a / np.sqrt(np.dot(a.H, a))
    
class BeamformedMonopoleArray(GeometryMixin, ClassicalBeamformerMixin, MonopoleElement):
    def nothing(self):
        pass
    
#Create beamformed array
arr = BeamformedMonopoleArray(4, np.pi / 4)
#arr = BeamformedMonopoleArray(4)
arr.plot_geometry()
plt.figure()
arr.plot_beam()
plt.figure()
arr.plot_gain()
```
