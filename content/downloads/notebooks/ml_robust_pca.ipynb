```python
import numpy as np
from numpy.linalg import norm, svd

def inexact_augmented_lagrange_multiplier(D, lmbda, tol=1e-7, maxiter=1000, verbose=True):
    """
    Inexact Augmented Lagrange Multiplier
    """
    Y = D
    norm_two = norm(Y.ravel(), 2)
    norm_inf = norm(Y.ravel(), np.inf) / lmbda
    dual_norm = np.max([norm_two, norm_inf])
    Y = Y / dual_norm
    A = np.zeros(Y.shape)
    E = np.zeros(Y.shape)
    dnorm = norm(D)
    mu = 1.25/norm_two
    rho = 1.5
    sv = 10.
    n = Y.shape[0]
    i = 0
    while True:
        Eraw = D - A + (1/mu) * Y
        Eupdate = np.maximum(Eraw - lmbda/mu, 0) + np.minimum(Eraw + lmbda / mu, 0)
        U, S, V = svd(D - Eupdate + (1 / mu) * Y, full_matrices=False)
        svp = (S > 1/mu).shape[0]
        if svp < sv:
            sv = np.min([svp + 1, n])
        else:
            sv = np.min([svp + round(.05 * n), n])
        Aupdate = np.dot(np.dot(U[:, :svp], np.diag(S[:svp] - 1/mu)), V[:svp, :])
        A = Aupdate
        E = Eupdate
        Z = D - A - E
        Y = Y + mu * Z
        mu = np.min([mu * rho, mu * 1e7])
        i += 1
        if (norm(Z, 'fro') / dnorm) < tol:
            if verbose:
                print "Converged at iteration %d" % (i)
            break
        elif i >= maxiter:
            if verbose:
                print "Maximum iterations reached, %d" % (i)
            break
    return A, E
```

```python
import numpy as np
from numpy.linalg import qr, norm

def wthresh(a, thresh, t='s'):
    res = np.abs(a) - thresh
    return np.sign(a) * ((res > 0) * res)

def go_dec(X, thresh=8, rank=2, power=10, max_iter=1000, error_bound=1e-6, random_seed=0):
    m, n = X.shape
    if m<n:
        X=X.T
    L=X
    S=np.zeros(L.shape)
    itr = 0
    random_state = np.random.RandomState(random_seed)
    while True:
        Y2 = random_state.randn(n, rank)
        for i in range(power):
            Y1=np.dot(L,Y2)
            Y2=np.dot(L.T,Y1);
        Q,R = qr(Y2,'full')
        L_new=np.dot(np.dot(L,Q), Q.T)
        T=L-L_new+S
        L=L_new
        S=wthresh(T, thresh)
        T-=S
        err=norm(T.ravel())
        if (err < error_bound) or (itr >= max_iter):
            break
        L+=T
        itr += 1
    G = X-L-S
    if m<n:
        L=L.T
        S=S.T
        G=G.T
    return L,S,G
```

```python
import urllib2
from scipy.io import loadmat, savemat
from scipy.misc import imresize
import os
import numpy as np
from numpy.linalg import svd

videoFname = "escalator_data.mat"

def download_video_clip():
    webFile = urllib2.urlopen(
        r'http://cvxr.com/tfocs/demos/rpca/escalator_data.mat')
    filename = './' + videoFname
    localFile = open(filename, 'w')
    localFile.write(webFile.read())
    localFile.close()

if not os.path.exists('./' + videoFname):
    print 'Video file not found, downloading'
    download_video_clip()
    
X = loadmat(videoFname)['X'].astype(np.float)/255.
d1, d2 = (160, 130)
nclip = X.shape[1]
lmbda = .01
```

```python
A, E = inexact_augmented_lagrange_multiplier(X, lmbda=lmbda, maxiter=100, verbose=0)
A = A.reshape(d1, d2, X.shape[1]).swapaxes(0, 1) * 255.
E = E.reshape(d1, d2, X.shape[1]).swapaxes(0, 1) * 255.
savemat("./IALM_background_subtraction.mat", {"A": A, "E": E})
print("RPCA complete")
```

```python
L, S, G = go_dec(X)
L = L.reshape(d1, d2, X.shape[1]).swapaxes(0, 1) * 255.
S = S.reshape(d1, d2, X.shape[1]).swapaxes(0, 1) * 255.
G = G.reshape(d1, d2, X.shape[1]).swapaxes(0, 1) * 255.
savemat("./GoDec_background_subtraction.mat", {"L": L, "S": S, "G": G})
print("GoDec complete")
```
